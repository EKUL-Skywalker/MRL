```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
csv = pd.read_csv("PLOTS - Sheet14.csv")
```

```python
csv.head()
```

```python
import numpy as np

names = {
    "R50-NRL": "MRL",
    "R50-NRL-E": "MRL-E",
    "R50-FF": "FF",
    "R50-SVD": "SVD",
    "SLIM": "Slim. Net",
    "R50-Rand": "Rand. FS"
}
sns.set_context('talk')
sns.set_style("white")
linestyles = ['-', '--', '-.', ':', "--", "-", ":", "-."]
markers = ['o', '^', 'v', 'h', '+', 'x', 'D', '*']

for i, ((name, mapping), linestyle, m) in enumerate(zip(names.items(), linestyles, markers)):
    print(name)
    if name == "R50-NRL":
        zorder = 30
    else:
        zorder = 2
    plt.plot(
        csv["Representation Size"], csv[name], label=mapping, marker=m, linewidth=3, linestyle=linestyle, zorder=zorder
    )

plt.ylim([40, 75])
plt.xscale("log")
plt.xticks(csv["Representation Size"], csv["Representation Size"], rotation=45)
plt.xlabel("Representation Size")
plt.ylabel("1-NN Accuracy (%)")
plt.grid()
plt.legend(loc=4)

plt.savefig("knn_classification-dim-x-acc.pdf", bbox_inches="tight")
```
