```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
csv = pd.read_csv("PLOTS - Sheet6.csv")
csv2 = pd.read_csv("PLOTS - Sheet7.csv")
```

```python
csv2
```

```python
tableau_20 = [
"k",
"tab:orange",
"tab:red",
"tab:green",
"tab:purple",
"tab:brown",
"tab:pink",
"tab:gray",
"tab:olive",
"tab:cyan"
]
```

```python
f, ax = plt.subplots(figsize=(16, 12))

sns.set_context('talk')
sns.set_style("white")
plt.ylim([55.75, 56.25])
plt.xlim([30, 3000])

markers = ['o', '^', 'v', 'h', '+', 'x', 'p', '*', 'D']
c=-1
for i in range(9):
    cmap = tableau_20[i]
    for j in range(i, 9):
        c+=1      
        m = markers[j]
        alpha=np.exp((csv["MAP@10"][c]-45))/(5*np.exp(10))
        if csv["MAP@10"][c]>56:
            alpha=0.8
        if csv["MAP@10"][c]>56.15:
            alpha=1
        plt.scatter(csv["MFLOPS"][c], csv["MAP@10"][c], marker=m, linewidth=3, linestyle='-', zorder=30, color=cmap, alpha=alpha, sizes=[5*csv["MAP@10"][c]])
fs=50
plt.yticks(fontsize=fs)
plt.xticks(fontsize=fs)
plt.xscale("log")
plt.xlabel("MFLOPS/Query", fontsize=fs)
plt.ylabel("MAP@10", fontsize=fs)
plt.grid()

for c in range(len(csv2["MFLOPS"])):
    plt.scatter(csv2["MFLOPS"][c], csv2["MAP@10"][c], marker='1', zorder=30, sizes=[800], c='tab:blue', linewidth=4)

for i in range(9):
    plt.scatter([], [],  marker=markers[i], label=f"{2**(i+3)}", c='steelblue')

legend1 = plt.legend(loc=4, borderpad=0.5, title="$D_{s}$", prop={'size': 20})
plt.gca().add_artist(legend1)

foo=[]; lab=[f"{2**(i+3)}" for i in range(9)]
for i in range(9):
    foo.append(plt.scatter([], [],  marker="s", c=tableau_20[i]))
    
legend2 = plt.legend(foo, lab, loc=1, borderpad=0.5, bbox_to_anchor=(0.86, 0.472), title="$D_{r}$", prop={'size': 20})    
plt.gca().add_artist(legend2)

foo=[]; lab=["Naive"]
foo.append(plt.scatter([], [],  marker="1", c="steelblue",linewidths=1))
legend2 = plt.legend(foo, lab, loc=1, borderpad=0.5, bbox_to_anchor=(0.925, 0.535), prop={'size': 20})    


plt.gca().add_artist(legend2)
plt.savefig("MAP_1k.pdf", bbox_inches="tight")
```

```python
import matplotlib

cmap = matplotlib.cm.get_cmap('BuPu')

rgba = [cmap(i) for i in np.linspace(0.1, 1.0, 9)]
```

```python
# f, ax = plt.subplots(figsize=(16, 12))

sns.set_context('talk')
sns.set_style("white")
plt.ylim([55.75, 56.25])
plt.xlim([30, 3000])

markers = ['o', '^', 'v', 'h', '+', 'x', 'p', '*', 'D']
c=-1
for i in range(9):
    cmap = rgba[i]
    for j in range(i, 9):
        c+=1      
        m = markers[j]
        alpha=np.exp((csv["MAP@10"][c]-9))/(np.exp(5.5))
        if csv["MAP@10"][c]>14.1:
            alpha=0.8
        if csv["MAP@10"][c]>14.3:
            alpha=1
        plt.scatter(csv["MFLOPS"][c],
                    csv["MAP@10"][c],
                    marker='o',
                    linewidth=3,
                    linestyle='-',
                    zorder=30, color=cmap, alpha=alpha, sizes=[1.75**j*2])

# fs=40
# plt.yticks(fontsize=fs)
# plt.xticks(fontsize=fs)
plt.xscale("log")
plt.xlabel("MFLOPS/Query")
plt.ylabel("MAP@10 (%)")
plt.grid()

for c in range(len(csv2["MFLOPS"])):
    plt.scatter(csv2["MFLOPS"][c], csv2["MAP@10"][c], marker='1', zorder=30, sizes=[200], c='r', linewidth=3)

for i in range(9):
    plt.scatter([], [],  marker='o', label=f"{2**(i+3)}", c='steelblue', sizes=[1.75**i*2])

legend1 = plt.legend(loc=1, borderpad=0.2, bbox_to_anchor=(1.545, 1.04),title="$D_{r}$", prop={'size':15})
plt.gca().add_artist(legend1)

foo=[]; lab=[f"{2**(i+3)}" for i in range(9)]
for i in range(9):
    foo.append(plt.scatter([], [],  marker="o", color=rgba[i]))
    
legend2 = plt.legend(foo, lab, loc=1, borderpad=0.2, bbox_to_anchor=(1.3, 1.04), title="$D_{s}$", prop={'size':15})    
# title = legend2.get_title()
# title.set_fontsize('x-large')
plt.gca().add_artist(legend2)

foo=[]; lab=["Naive"]
foo.append(plt.scatter([], [],  marker="1", c="r"))
legend2 = plt.legend(foo, lab, loc=1, borderpad=0.5, bbox_to_anchor=(1, 0.21))    


# plt.gca().add_artist(legend2)
plt.savefig("MAP_1K.pdf", bbox_inches="tight")
```
